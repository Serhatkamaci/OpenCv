## 5 çemberden oluşan yapıyı trackbar ile  sağa, sola ve yukarı, aşağı kaydırma.
## It moves the structure consisting of 5 circles to the right, left, up and down with the movement bar.

```python
def function(x):
    pass

cv2.namedWindow("Screen",cv2.WINDOW_NORMAL)

# Trackbar Oluşumu
cv2.createTrackbar("Left-Right","Screen",0,520,function)
cv2.createTrackbar("Up-Down","Screen",0,520,function)

left_right_position_merkez_nokta_1_2=256
up_down_position_merkez_nokta_1_2=256

yaricap_artis_noktasi=35

# Ben left_right veya up_down işlemleri için bir nokta kaydıracağım sadece çember oranlarını korumak için 
# "yaricap_artis_noktasi" dahil edeceğim.

while True:
    
    screen=np.zeros((512,512,3),np.uint8)+255

# Circle oluşumu

# 2 nokta arası uzaklığı kullanıp diğer çemberlerin hepsini merkez noktaları üzerinden geçireceğiz.
# Yarıçap 50 olduğu için tahmini olarak x ve y değerlerini farklı çemberler için 35,333 falan geliyor ancak 35 olarak kaydırma işlemleri 
# gerçekleştirdim.

    cv2.circle(screen,(left_right_position_merkez_nokta_1_2,up_down_position_merkez_nokta_1_2),50,(0,0,0),thickness=2) # Merkez circle
    cv2.circle(screen,(left_right_position_merkez_nokta_1_2-35,up_down_position_merkez_nokta_1_2-35),50,(255,0,0),thickness=2) # Sol üst circle
    cv2.circle(screen,(left_right_position_merkez_nokta_1_2+35,up_down_position_merkez_nokta_1_2-35),50,(0,255,0),thickness=2)# Sağ üst circle
    cv2.circle(screen,(left_right_position_merkez_nokta_1_2-35,up_down_position_merkez_nokta_1_2+35),50,(0,0,255),thickness=2) # Sol alt circle
    cv2.circle(screen,(left_right_position_merkez_nokta_1_2+35,up_down_position_merkez_nokta_1_2+35),50,(255,0,255),thickness=2) # Sağ alt circle


    cv2.imshow("Screen",screen)        
        
    left_right_position_merkez_nokta_1_2= cv2.getTrackbarPos("Left-Right","Screen")
    up_down_position_merkez_nokta_1_2=cv2.getTrackbarPos("Up-Down","Screen")
    
    
    if cv2.waitKey(1) & 0XFF==ord("q"):
        break
        
cv2.destroyAllWindows()
```

## OpenCV ile boxplot çizimi.
## Boxplot drawing with OpenCV.

```python
# Sayı doğrusu artış oranımız 5 olsun.

minVal=10
maxVal=70
q1=25
q3=45
median=30

# Kutu grafiği çizerken bir referans noktası vermemiz gerekiyor. Başlangıç konumu, katsayı için bir değer.

offsethorizontal=50  # Yataydaki konum.
offsetvertical=400 # Dikeydeki konum.
coefficient=5
tut_horizontal_1= offsethorizontal# Aşağıda yaptığım her değişikliklerde horizontal değerim değişecek.

cv2.namedWindow("Screen",cv2.WINDOW_NORMAL)
screen=np.zeros((500,500,3),dtype=np.uint8)+255

# Sayı doğrusu 

sayi_dogrusu_uzunluk=400

cv2.line(screen,(offsethorizontal,offsetvertical),(450,offsetvertical),(0,0,0),2)

# Dikey Çizgileri Çiz

# Metinlerin referans noktası (50,400) olacak. Artış miktarımız ise 5 olsun.


for i in range(0,410,25): # Son noktayı görmediği için 10 ekleyip diğer noktayı yazrımadım.Sayı doğrusu uzunluğu zaten 400'idi.
    
    cv2.line(screen,(offsethorizontal,offsetvertical+10),(offsethorizontal,offsetvertical),(0,0,0),1)
    offsethorizontal+=25
    
offsethorizontal=tut_horizontal_1-2  # Eski değerine geri getirdim.
    
# Sayı doğrusunu doldur    

# Dikey pozisyon dikey çizimlerinde 10 birimdi bunu 15 birime getirip x leri arttırma işlemleri gerçekleştirelim.
yeni_vertical=offsetvertical+20

for i in range(0,410,25):
    
    cv2.putText(screen,str(i),(offsethorizontal-2,yeni_vertical),cv2.FONT_HERSHEY_DUPLEX,0.3,(0,0,0))  
    offsethorizontal+=25
    
offsethorizontal=tut_horizontal_1


# Min-q1 arası

cv2.line(screen,(offsethorizontal+minVal*coefficient,offsetvertical-50),(offsethorizontal+q1*coefficient,offsetvertical-50),(0,0,0),2)
cv2.putText(screen,"Min",(offsethorizontal+minVal*coefficient-5,offsetvertical-60),cv2.FONT_ITALIC,0.3,(0,0,0))

# Kutu çizimi

cv2.rectangle(screen,((offsethorizontal+q1*coefficient), offsetvertical-75),(offsethorizontal+q3*coefficient,offsetvertical-25),(0,0,0),2)
cv2.putText(screen,"q1",((offsethorizontal+q1*coefficient-5), offsetvertical-85),cv2.FONT_ITALIC,0.3,(0,0,0))

# Median 

cv2.line(screen,(offsethorizontal+median*coefficient,offsetvertical-75),(offsethorizontal+median*coefficient,offsetvertical-25),(0,255,0),2)
cv2.putText(screen,"Median",(offsethorizontal+median*coefficient-14,offsetvertical-85),cv2.FONT_ITALIC,0.3,(0,0,0))

# Median-q3 arası

cv2.line(screen,(offsethorizontal+q3*coefficient,offsetvertical-50),(offsethorizontal+maxVal*coefficient,offsetvertical-50),(0,0,0),2)
cv2.putText(screen,"q3",(offsethorizontal+q3*coefficient-5,offsetvertical-85),cv2.FONT_ITALIC,0.3,(0,0,0))
cv2.putText(screen,"Max",(offsethorizontal+maxVal*coefficient-5,offsetvertical-60),cv2.FONT_ITALIC,0.3,(0,0,0))

cv2.imshow("Screen",screen)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
